```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
```

```python
X = cr_cd.drop('Class', axis=1)
y = cr_cd['Class']
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
print("Before: ", y.value_counts())
print("After: ", y_resampled.value_counts())
#now our class is balanced, which helps the model learn better 
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_resampled[['Amount', 'Time']] = scaler.fit_transform(X_resampled[['Amount', 'Time']])
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled)
```

```python
X_train
```

```python
#import libraries
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
#LogisticRegression
lr = LogisticRegression(max_iter = 1000)
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# create Random Forest classifier
# Using all cores for faster training, and 100 trees for balance between speed and accuracy
rf = RandomForestClassifier(n_estimators=100, n_jobs=-1, random_state=42)

# train the model
rf.fit(X_train, y_train)

#predict on test data
y_pred_rf = rf.predict(X_test)
```

```python
print(">Accuracy:", accuracy_score(y_test, y_pred_rf))
print("\n >Classification Report:\n", classification_report(y_test, y_pred_rf))
print("\n >Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(y_test, y_pred_rf, cmap='Blues')
plt.title("Random Forest Confusion Matrix")
plt.show()
```

```python
from sklearn.metrics import roc_auc_score, roc_curve

# Predict probabilities (not just labels)
y_probs_rf = rf.predict_proba(X_test)[:, 1]  # Probabilities for class 1

fpr, tpr, thresholds = roc_curve(y_test, y_probs_rf)
roc_auc = roc_auc_score(y_test, y_probs_rf)

# Plot ROC Curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.2f})', color='darkorange')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Random Forest')
plt.legend()
plt.grid(True)
plt.show()
```
